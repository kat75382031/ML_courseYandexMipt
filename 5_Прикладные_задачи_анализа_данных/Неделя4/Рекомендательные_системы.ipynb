```python
def readFile(file):
    viewed=[]
    purchased=[]
    ids_viewed=[]
    ids_purchased=[]
    f=open(file, 'r')
    for line in f:
        elements = line.rstrip().split(";")
        viewed.append(elements[0].split(','))
        purchased.append(elements[1].split(','))
        for el in elements[0].split(','):
            if el!='':
                ids_viewed.append(el)
        for el in elements[1].split(','):
            if el!='':
                ids_purchased.append(el)
    return viewed,purchased,ids_viewed,ids_purchased
```

```python
def precision_k(rec,purchased):
    count=0
    for i in range(len(rec)):
        for j in range(len(purchased)):
            if rec[i]==purchased[j]:
                count+=1
    return count/len(rec)
```

```python
def recall_k(rec,purchased):
    count=0
    for i in range(len(rec)):
        for j in range(len(purchased)):
            if rec[i]==purchased[j]:
                count+=1
    return count/len(purchased)
```

```python
def MAP_k(viewed,purchased,k,rec):
    precision=[]
    for i in range(len(viewed)):
        if purchased[i][0]!='':
            if k>len(viewed[i]):
                precision.append(AP_k(rec[-len(viewed[i]):],purchased[i]))
                print(len(viewed[i]))
            else:
                precision.append(AP_k(rec[-k:],purchased[i]))
                print(k,len(viewed[i]))
                
    print(precision)                                            
    return np.mean(precision)
```

```python
def AR_k(viewed,purchased,k,rec):
    recall=[]
    for i in range(len(viewed)):
        if purchased[i][0]!='':
            if k>len(viewed[i]):
                recall.append(recall_k(rec[-len(viewed[i]):],purchased[i]))
            else:
                recall.append(recall_k(rec[-k:],purchased[i]))
    return sum(recall)/len(recall)
```

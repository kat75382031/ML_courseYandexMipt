```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import sqrt
%matplotlib inline
```

```python
k = 4.
pareto_rv=sts.pareto(k) # задаем случайную величину, распределенную по Парето
rv=pareto_rv.rvs(size=1000) # сгенерируем 1000 значений
x = np.linspace(1,5, 1000)
pdf=pareto_rv.pdf(x)
```

```python
plt.hist(rv,bins=20,density=True) # строим гистограмму выборки
plt.plot(x,pdf) #  и теоретическую плотность распределения
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
EX=k/(k-1) # математическое ожидание СВ, распределенной по Парето
DX=((1./(k-1))**2)*k/(k-2) # дисперсия СВ, распределенной по Парето
```

```python
# генерируем 1000 выборок размером 5, для каждой считаем выборочное среднее и записываем в массив полученое значение
a1=np.zeros(1000)
for j in range(1000):
    rv = pareto_rv.rvs(size=5)
    for k in range(5):
        a1[j]+=rv[k]
    a1[j]/=5.
# задаем нормально распределенную случайную величину, аппроксимирующую распределение выюорочных средних при n=5
norm_rv=sts.norm(loc=EX,scale=sqrt(DX/5))
x=np.linspace(0,2,100)
# строим гистограмму выборочных средних и плотность соответствущего нормального распределения
plt.hist(a1,bins=30,density=True)
plt.plot(x,norm_rv.pdf(x))
plt.ylabel('$f(x)$')
```

```python
# генерируем 1000 выборок размером 10, для каждой считаем выборочное среднее и записываем в массив полученое значение
a1=np.zeros(1000)
for j in range(1000):
    rv = pareto_rv.rvs(size=10)
    for k in range(10):
        a1[j]+=rv[k]
    a1[j]/=10.
# задаем нормально распределенную случайную величину, аппроксимирующую распределение выюорочных средних при n=10
norm_rv=sts.norm(loc=EX,scale=sqrt(DX/10))
x=np.linspace(0.9,2,100)
# строим гистограмму выборочных средних и плотность соответствущего нормального распределения
plt.hist(a1,bins=30,density=True)
plt.plot(x,norm_rv.pdf(x))
plt.ylabel('$f(x)$')
```

Как видно, распределение Парето не позволяет распределению выборочных средних быстро сходиться к нормальному ввиду его скошенности. Из графиков можно заметить, что нормальное распределение начинает неплохо аппроксимировать распределение выборочных средних при размерах выборки, не меньших чем 50

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

```python
#посчитаем среднее E и дисперсию D
Xm = 1
k = 4 # k == b == коэффициент кривизны
E = (Xm * k) / (k - 1)
D = (E**2)*(k/(k-2))
print('Серднее = ' +str(E) + '; ' 'Дисперсия = ' + str(D))
```

```python
#задаем функцию
pareto = sts.pareto(b = 4)
#генерируем выборку необходимого размера
sample = pareto.rvs(1000)
x = np.linspace(1,5,1000)
pdf = pareto.pdf(x)
#строим гистограмму выборки и теоретических значений
plt.plot(x, pdf, label = 'Theoretical PDF')
plt.hist(sample, bins = 100, range = (1, 5), density = True)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend(loc = 'best')
```

```python
#создаем функцию для ЦПТ, где n число выборок
def func(n):
    sample = []
    for i in range(1000):
        sample_i = pareto.rvs(n)
        sample.append(sum(sample_i)/n)
    D_i = D / n
    sigma = math.sqrt(D_i)
    norm = sts.norm(E, sigma)
    pdf = norm.pdf(x)
    plt.plot(x, pdf, label='PDF (n = {})'.format(n))
    plt.hist(sample, bins = 30, range=(1, 3), density = True)
    plt.ylabel('f(x)')
    plt.xlabel('x')
    plt.legend(loc='best')
```

```python
#функция для объема выборки = 5
func(5)
```

```python
#функция для объема выборки = 10
func(10)
```

```python
#функция для объема выборки = 50
func(5000)
```

Вывод:
С увеличением объема выборки функция становится более похожей на нормальное распределние.

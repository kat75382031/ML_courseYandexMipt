```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#Распределение Максвелла
a=2
raspr = sts.maxwell(scale=a)

#Теоретическая плотность распределения случайной величины:
x = np.linspace(-1,10,50)
pdf = raspr.pdf(x)

#Графики выборки и плотности распределения
plt.plot(x, pdf, label='Функция распределения Максвелла')

#Расчет мат.ожидания + дисперсии для 
mu=2*a*math.sqrt(2/math.pi)
si2=a*a*(3*math.pi-8)/math.pi

norm_rv = sts.norm(mu, math.sqrt(si2))
pdf_mean = norm_rv.pdf(x)
plt.plot(x, pdf_mean, label='Функция норм.распр. при тех же Мат.Ож + Дисп.')
plt.legend()
```

```python
#Выборка из n=5
n=5
sets = list(range(1,1001))

emv=[]

for i in sets:
    sample = raspr.rvs(size=n) 
    sum=.0
    for j in sample:
        sum=sum+j
    emv.append(sum/n)    

norm_rv = sts.norm(mu, math.sqrt(si2/n))
pdf_mean = norm_rv.pdf(x)

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('x')
ax1.set_ylabel('num', color=color)
ax1.hist(emv, bins=30, density=True) 

ax2 = ax1.twinx() 

color = 'tab:red'
ax2.set_ylabel('f(x)', color=color) 
ax2.plot(x, pdf_mean, color=color, label='Функция норм.распр. при n=5')

plt.legend()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import math
from matplotlib import pylab as plt
```

В качестве искомого вектора х выступает векто весов w

```python
def f(x):
    return math.sin(x / 5.) * math.exp(x / 10.) + 5 * math.exp(-x / 2.)
print(f(5))
```

## Аппроксимация многочленом первой степени

```python
A=np.array([[1,1], [1,15]])
b=[f(1),f(15)]
```

```python
w = np.linalg.solve(A,b)
```

```python
A.dot(w) # проверка
```

```python
x=np.arange(1,16,1)
y=[]
for i in range(15):
    y.append(f(i+1))
plt.plot([1,15],[w[0]+w[1],w[0]+w[1]*15])
plt.plot(x, y)
```

## Многочлен второй степени

```python
A=np.array([[1,1,1],[1,8,8*8],[1,15,15*15]])
b=[f(1),f(8),f(15)]
```

```python
w = np.linalg.solve(A,b)
```

```python
y2=[]
for i in range(15):
    y2.append(w[0]+w[1]*x[i]+w[2]*x[i]*x[i])  
plt.plot(x,y2)
plt.plot(x,y)
```
